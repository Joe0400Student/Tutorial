```python
"""
Written By joe0400 on reddit

Tutorial written in open domain,
    to help teach people programming
    a usefull helpful skill, on
    a beuatiful language.
    
    Only request I have, is for
    you to share and spread this
    where needed. And if this
    helped you, let me know. Its
    always nice to see that.
"""


    
    
    #
    #  This is a really complicated list comprehensions, list comprehensions work
    #  by generating values and are set up like this for 1 length list comprehensions
    #  for two length list comprehensions (two if statements or more), they are done
    #  like this [ i if bool else j if bool else k for generator ]
    #
    
def prime_comprehension(floor: int , ceil: int): return [i for i in range(floor , ceil) if not sum([i%j==0 for j in range(2,i)])]
    
    #
    #  Generators are a type of method that is vastly quicker at some functions. The good
    #  thing about generators, is they reduce memory usage often, due to forgetting large
    #  previous states that returning lists cause. this is why for example the itertools
    #  library is written like this. if you had to find a certain set of permutations of
    #  50 bits, it would end up being ~ 1 quadrillion long (2^50-1) list in binary notation
    #  Generators has certain speed advantages as well
    #  Generators will always return to the last yeild and run from there
    #  for statements will always end before the generator throws a exception, specifically
    #  the StopIteration exception
    #
    
def prime_generator(ceil: int):
    
    prime_list = []
    
    for i in range(2 , ceil):
        
        if not len(prime_list):
            prime_list.append(i)
            yield i
        else:
            check = True
            for j in prime_list:
                if i % j == 0:
                    check = False
                    break
            if check:
                prime_list.append(i)
                yield i
        
    #
    #  Finally, there is something called a lambda expression
    #  Lambda expressions defined in PEP 8, is different then
    #  how most people use them in python. 
    #  They can be either
    #     sqrt = lambda n : n ** 2  #Common style
    #     def sqrt(n): return n ** 2 #PEP 8 Standard
    #  For examples sake, I will put both of them up
    #
    
lambda_sqr = lambda n : n ** 2
    
def pep_lambda_sqr(n): return n ** 2
    

    #
    #  Their is also something called dictionary comprehension
    #  If your interested, read up on it. I find it less usefull
    #  Because everything a dictionary could do, a list can do.
    #
    

print("Prime Comprehension: " , end='')
print(prime_comprehension(2,200))
print("Prime Generator = [" , end='')
for i in prime_generator(200):
    print(i,end=", ")
print(']')
print("Lambdas : ")
print(lambda_sqr(20))
print(pep_lambda_sqr(20))
```
